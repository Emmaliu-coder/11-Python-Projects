## 4.3 集合 { }

新增一空集合

````
S = set()
print(S)
````

````
len(S)
````

把一個元素加進一個集合裡(一次只能加入一個值)

````
S.add(2)
print(S)
````

把資料存成集合，會讓重複的值只出現一次

````
S1 = set(b)
print(b)
print(S1)
````

````
S2 = set(a)
S2.add(-2)
print(a)
print(S2)
````

### 交集




````
S1.intersection(S2)
````

(可交集多個集合:.intersection().intersection())

````
a = {1,2,3}
b = {1,5,6}
c = {1,7,8}
a.intersection(b).intersection(c)
````
